```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
import plotly.express as px
from sklearn.metrics import accuracy_score

cal_housing = fetch_california_housing() 
X = pd.DataFrame(cal_housing.data, columns=cal_housing.feature_names) 
y = cal_housing.target 
x = X['MedInc']
df = pd.DataFrame( 
        dict(MedInc=x, Price=cal_housing.target))
y = np.array(y)
y
```

```python
fig = px.scatter(df, x=x, y=y )
fig.show()
```

```python
loss, b0, b1 = [], [], []
for i, _b0 in enumerate(np.linspace(-100, 100, 100)):
    if i == 80:
         for _b1 in np.linspace(-100, 100, 100):
            b0.append(_b0)
            b1.append(_b1)
            loss.append(np.power((y - (_b1 * x + _b0)), 2).max())
l = pd.DataFrame(dict(b0=b0, b1=b1, loss=loss))
fig = px.scatter(l, x="b1", y="loss")
fig.show()
```

```python
loss, b0, b1 = [], [], []
lam = 0.1
for i, _b0 in enumerate(np.linspace(-100, 100, 100)):
         for _b1 in np.linspace(-100, 100, 100):
            b0.append(_b0)
            b1.append(_b1)
            loss.append(np.power((y - _b1 * x - _b0), 2).max() + lam * ( 2 ** _b0 + 2 ** _b1 ))
regloss = pd.DataFrame(dict(b0=b0, b1=b1, loss=loss))
fig = px.density_contour(regloss, x="b0" ,y="b1", z="loss")
fig.show()
```

```python
print("starting sgd")
beta = np.random.random(2)
loss, b0, b1 = [], [], []
alpha = 0.001
lam = 0.1

for i in range(1000):
        y_pred: np.ndarray = beta[0] + beta[1] * x

        g_b0 = -1 * (y - y_pred).max() + 2 * lam * beta[0]
        g_b1 = -1 * (x * (y - y_pred)).max() + 2 * lam * beta[1]

        print(f"({i}) beta: {beta}, gradient: {g_b0} {g_b1}")

        beta_prev = np.copy(beta)

        beta[0] = beta[0] - alpha * g_b0
        beta[1] = beta[1] - alpha * g_b1

        if np.linalg.norm(beta - beta_prev) < 0.001:
            print(f"I do early stoping at iteration {i}")
            break
```
